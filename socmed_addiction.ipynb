**Project Background**

Nowadays, social media platforms have changed from just a communication tool to platforms that strongly influence daily human activity. Platforms such as Instagram, TikTok, and WhatsApp has engagement features like personalized content and constant notification that makes the users tend to constantly stay on their screen for a long period of time. As a result, teenagers, especially students in schools become more vulnerable to addictive online behaviour. Although social media provides many benefits such as global communication, and educational sources, uncontrolled usage has become a growing concern around the world. This lifestyle became an addiction and drives the user to excessively immerse in their screen and disrupt their normal sleep habits. The common habits of using screens before sleeping exposes users to blue light and increased mental stimulation, which studies have linked to shorter sleeping hour (Alshoaibi et al., 2023; Mohd Saat et al., 2024). 

Aside from that, excessive social media usage also changes interpersonal relationships and social behaviour. As online engagement increases, face-to-face social interaction often decreases significantly, causing students to move away from meaningful real-world relationships and increasing feelings of social isolation. In addition, the combined effects of sleep deprivation and reduced social interaction may eventually affect students’ academic performance, such as CGPA which is student’s Cumulative Grade Point Average. 

To address these growing lifestyle concerns, research-based and data-driven findings are needed to clearly identify how excessive digital usage leads to behavioural problems. This study uses an anonymous student dataset that records digital activity together with lifestyle outcomes, including addiction levels, night-time screen habits, sleep duration, and real-world social interaction indicators. By examining these connected variables, this mini project aims to identify the point at which normal social media use becomes harmful dependency, while also providing useful insights for academic counsellors, students, and educational policymakers who aim to restore healthy digital and academic balance. 

**Project Objective**

- To illustrate the percentage age distribution of teenagers across different addiction categories (Low, Medium, High).
- To measure and examine teenagers social media addiction level and their daily social media usage per hour.
- To examine the relationship between social media addiction and physiological well-being indicators, specially on students’ sleep duration. 
- To investigate the consequences of excessive digital consumption by analysing relationship between a student’s social media addiction and their social interaction level.


**Data Loading**

```python
import pandas as pd
import numpy as np

data = pd.read_csv("Social_Media_Addiction.csv")
print(data)
```

```python
#Load data into DataFrame object
df = pd.DataFrame(data)
```

**Data Preparation**

```python
#Renaming columns
df = df.rename(columns={'age': 'Age', 'gender': 'Gender', 'daily_social_media_hours': 'Daily Social Media per hour', 'platform_usage': 'Platform Usage', 'sleep_hours': 'Sleep Hours', 'screen_time_before_sleep': 'Screen time before sleep', 'physical_activity': 'Physical Activity', 'social_interaction_level': 'Social Interaction Level', 'addiction_level': 'Addiction Level'})
df
```

```python
#Check for missing values
df.isnull().sum()
```

```python
#Drop rows with missing values
df2 = df.dropna()

#Check for missing values again
df2.isnull().sum()
```

```python
#Identify duplicates
df2.duplicated().sum()
```

```python
# Show the duplicate rows
df2[df2.duplicated(keep=False)]
```

```python
#Remove duplicate rows
df3 = df2.drop([1200, 1201, 1202])
```

```python
#Check for duplicates
df3.duplicated().sum()
```

```python
df3.info()
```

```python
# make a copy so df4 is independent
df4 = df3.copy()  

#Add column
df4["Addiction Category"] = pd.cut(
    df4["Addiction Level"],
    bins=[0, 3, 6, 9], # ranges: 0–3, 3–6, 6–9
    labels=["Low", "Medium", "High"],
    include_lowest=True
)

df4
```

```python
#Change data type
df4['Age'] = df4['Age'].astype('int')
df4.info()
```

```python
#Save as .csv file
df4.to_csv('socmed_cleaned.csv', index=False)
```

```python
#Load data
DF = pd.read_csv('socmed_cleaned.csv')
DF.info()
```

**Data Visualization**

```python
import matplotlib.pyplot as plt

# Filter teenagers (age 13–19)
teenagers = DF[(DF["Age"] >= 13) & (DF["Age"] <= 19)]

# Count number of teenagers in each Addiction Category
counts = teenagers["Addiction Category"].value_counts().reset_index()
counts.columns = ["Addiction Category", "Count"]

# Explicitly set category order
order = ["Low", "Medium", "High"]
counts["Addiction Category"] = pd.Categorical(counts["Addiction Category"], categories=order, ordered=True)

# Plot pie chart
plt.figure(figsize=(7,7))
plt.pie(
    counts["Count"], 
    labels=counts["Addiction Category"], 
    autopct="%1.1f%%", 
    startangle=90, 
    colors=["#8fd9b6", "#f6c85f", "#ff6f61"]  # optional custom colors
)

plt.title("Percentage of Teenagers by Addiction Category")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.regplot(x="Daily Social Media per hour", y="Addiction Level", data=DF, scatter_kws={"alpha":0.6}, line_kws={"color":"red"})
plt.title("Relationship between Addiction Level and Daily Social Media Usage")
plt.xlabel("Daily Social Media per hour")
plt.ylabel("Addiction Level")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x="Sleep Hours", y="Addiction Level", data=DF, scatter_kws={"alpha":0.6}, line_kws={"color":"red"})
plt.title("Relationship between Addiction Level and Sleep Hours")
plt.xlabel("Sleep Hours")
plt.ylabel("Addiction Level")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 5))

#Create a bar plot
sns.barplot(
    data=DF, 
    x='Social Interaction Level', 
    y='Addiction Level', 
    hue='Social Interaction Level', # Explicitly sets the hue to match x
    order=['low', 'medium', 'high'],
    palette='Blues_r',
    #legend=False,                    # Prevents an unnecessary duplicate legend
    errorbar=None 
)

# Add labels
plt.title('Average Social Media Addiction Level by Social Interaction Category', 
          fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Real-World Social Interaction Level', fontsize=11)
plt.ylabel('Average Addiction Level Score', fontsize=11)

plt.tight_layout()
plt.show()
```

**Discussion**

The findings obtained from the dataset provide strong evidence that support the study’s initial objectives. By analysing the behavioural and physiological data of the teenager group, clear patterns emerge regarding how excessive social media use affects important aspects of students’ daily lives.

The first graph which is pie chart, visualize the percentage distribution of teenagers by addiction category. The graph shows that majority of teenagers fall into medium level of social media addiction, followed by a significant portion at the high level. Only a small portion of teenagers exhibit low social media addiction. This highlights a concerning trend where moderate to high levels of addiction dominate among the teenage group.

The second graph shows a regression plot illustrating the relationship between daily social media usage per hours and addiction level among teenagers. The points reveal a clear  upward trend, indicating that as the number of hours spent on social media increases, the addiction level also tends to rise. The red positive regression line highlight heavier social media use is strongly associated with higher addiction levels.

Addressing the third objective, another scatter plot was used to examine the relationship between social media addiction and sleep duration. The result of the plotting shows a clear negative linear correlation. As illustrated by the downward-sloping regression line, higher addiction levels consistently correspond with reduced nightly sleep hours. This finding strongly supports the theoretical framework discussed in the background of this study. The data suggests that compulsive social media use disrupts healthy sleep habits.

In addition, when students spend excessive amounts of time on digital platforms, the reduction in sleep is further worsened by factors such as nighttime blue-light exposure and increase mental stimulation before sleep (Alshoaibi et al., 2023; Mohd Saat et al., 2024). Instead of achieving the recommended 7 to 9 hours of sleep needed for a healthy body and psychological well-being, highly addicted students tend to cluster around shorter sleep durations, making them more vulnerable to digital fatigue.

Furthermore, for the second objective, 
Lastly, a bar chart was used to investigate the relationship between a students’ level of real-world social interaction and their average digital addiction score. The result of the plot reveals a clear step-down relationship. Students categorized within the “low’ social interaction group recorded the highest addiction scores (≈6.5), whereas those with “high” levels of real-world interaction shows significantly lower addiction scores (≈3.5).

Although the bar chart presents social interaction levels on the X-axis for comparison between groups, the real-world implications suggest a two-way relationship. Students who experience loneliness may turn to social media more frequently, but it is also likely that increasing social media addiction actively draws students away from face-to-face interactions, leading to a decline in their real--world social engagement.

This pattern supports the “time displacement” theory of behavioural addiction (Dienlin & Johannes, 2020). As virtual interactions occupy a larger portion of a students’ daily time and attention, real-world social interactions tend to decrease, contributing to social isolation and highlighting social media addiction as big threat to interpersonal development.

**Conclusion**

This research project aimed to investigate the various consequences of excessive digital consumption among students, focusing specially on its physiological and social impacts. By analysing empirical data using data science methods within Jupyter Notebooks framework, the study successfully achieved both of its main objectives. 

The investigation shows that student digital dependency does not exist on its own, as it actually also actively affects the important aspects of healthy development. What can be seen physically from the increasing social media addiction is the decline in nightly sleep duration, indicating screen use as a major factor that disrupts healthy sleep habits. Meanwhile in social aspect, a distinct pattern shows confirms that lower level of real-world social interaction is associated with higher digital addiction scores, supporting the theory of media time displacement. 

Ultimately, these findings suggest that social media addiction creates a cycle in which social isolation and sleep deprivation reinforce one another. In practice, this cycle develops when late-night social media use led to sleep deprivation, leaving students too tired to participate in face-to face social activities. As a result, isolation can increase the feeling of loneliness, which may encourage the students to spend even more time on digital platforms to reduce the boredom causing the cycle to continue. 

Therefore, to protect the well-being of students, these results indicate that educational institutions should play an active role in addressing excessive screen use. Providing programs that teach responsible technology use and promoting face-to-face activities can help students develop healthier technology habits and achieve a better balance between their digital and real-world lives.
